# IPL Data Analysis Assignment 


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
matches.head()
```

## 1. Match won by maximum runs

```python
max_run = matches.loc[matches['win_by_runs'].idxmax()]
print(max_run[['season','team1','team2','winner','win_by_runs']])
```

## 2. Match won by maximum wickets

```python
max_wicket = matches.loc[matches['win_by_wickets'].idxmax()]
print(max_wicket[['season','team1','team2','winner','win_by_wickets']])
```

## 3. Match won by minimum runs

```python
min_run_match = matches[matches['win_by_runs'] > 0]
min_run = min_run_match.loc[min_run_match['win_by_runs'].idxmin()]

print(min_run[['season','team1','team2','winner','win_by_runs']])
```

## 4. Match won by minimum wickets

```python
min_wicket_match = matches[matches['win_by_wickets'] > 0]
min_wicket = min_wicket_match.loc[min_wicket_match['win_by_wickets'].idxmin()]

print(min_wicket[['season','team1','team2','winner','win_by_wickets']])
```

## 5. Matches where D/L applied vs not applied

```python
dl = matches['dl_applied'].value_counts()
plt.bar(dl.index, dl.values , color=['blue', 'orange'])
plt.title("D/L Method Applied",color='red')
plt.xlabel("DL Applied (0=No,1=Yes)")
plt.ylabel("Matches")
plt.figtext(0.9, 0.01, "Deepesh Srivastava", horizontalalignment='right')
plt.show()
```

<b>INSIGHTS<b> <br>
<ol>
<li>Most IPL matches are completed without using the D/L method.</li>
<li>This shows that rain affects only a small number of matches.</li>
<ol>

## 6. Number of matches in each city

```python
city = matches['city'].value_counts().head(10)
plt.bar(city.index, city.values, color='green')
plt.xticks(rotation=40)
plt.title("Top Cities Hosting IPL Matches", color='red')
plt.xlabel("City")
plt.ylabel("Matches")
plt.figtext(0.9, 0.01, "Deepesh Srivastava", horizontalalignment='right')
plt.show()
```

<b>INSIGHTS<b> <br>
<ol>
<li>Mainly, most of the matches are held in Mumbai and Kolkata.</li>
<ol>

## 7. Matches won by each team

```python
team_wins = matches['winner'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(team_wins.index, team_wins.values, color=['blue', 'orange', 'green', 'red', 'purple', 'cyan', 'magenta', 'yellow'])
plt.xticks(rotation=90)
plt.title("Matches Won by Each Team")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.figtext(0.9, 0.01, "Deepesh Srivastava", horizontalalignment='right')
plt.show()
```

<b>INSIGHTS<b> <br>
<ol>
<li>Mumbai Indians have won the most matches in the IPL.</li>
<li>Top 2nd highest team is Chennai Super Kings</li>
<li>3ed position team is kolkata Knights Riders</li>
<ol>

## 8. Matches held every season

## 10. Does winning toss mean winning match

```python
matches['toss_match_win'] = matches['toss_winner'] == matches['winner']
result = matches['toss_match_win'].value_counts()
plt.pie(result.values, labels=['Lost','Won'], autopct='%1.1f%%',colors=['red','green'])
plt.title("Toss Winner vs Match Winner" ,color='red')
plt.figtext(0.9, 0.01, "Deepesh Srivastava", horizontalalignment='right')
plt.show()
```

<b>INSIGHTS<b> <br>
<ol>
<li>The result mainly depends on how well the team plays.</li>
<li>Toss gives a small advantage, but performance is more important.</li>
<ol>

## 11. Toss decision taken by captains

<b>INSIGHTS<b> <br>
<ol>
<li>Most captains prefer to field first after winning the toss.</li>

<ol>

## 12. D/L method applied every season
